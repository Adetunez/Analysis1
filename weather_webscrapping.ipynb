```python
def scrape_weather_data(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    FeelsLikeSection = soup.find('li', {'data-testid': 'FeelsLikeSection'}).text
    WindSection = soup.find('li', {'data-testid': 'WindSection'}).text
    HumiditySection = soup.find('li', {'data-testid': 'HumiditySection'}).text
    uvIndexSection = soup.find('li', {'data-testid': 'uvIndexSection'}).text
    CloudCoverSection = soup.find('li', {'data-testid': 'CloudCoverSection'}).text
    AccumulationSection = soup.find('li', {'data-testid': 'AccumulationSection'}).text

    # Process the extracted data as needed
    print(FeelsLikeSection, WindSection, HumiditySection, uvIndexSection, CloudCoverSection, AccumulationSection)

# Replace 'https://weather.com/en-NG/weather/hourbyhour/l/36a7da1656fdfd4979d4d74a65d97b068cb7b3d597b5ea254adfaaaf358d53ca' with the actual URL you want to scrape
url = "https://weather.com/en-NG/weather/hourbyhour/l/36a7da1656fdfd4979d4d74a65d97b068cb7b3d597b5ea254adfaaaf358d53ca"

scrape_weather_data(url)
```

```python
# Define variables outside the function
FeelsLikeSection = ""
WindSection = ""
HumiditySection = ""
uvIndexSection = ""
CloudCoverSection = ""
AccumulationSection = ""
```

```python
# Import the CSV library
import csv

# Open a CSV file for writing
with open('weather_data.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Feels Like', 'Wind', 'Humidity', 'UV Index', 'Cloud Cover', 'Accumulation'])  # Header row
    writer.writerow([FeelsLikeSection, WindSection, HumiditySection, uvIndexSection, CloudCoverSection, AccumulationSection])
```

```python
import pandas as pd

# Read the CSV file into a Pandas DataFrame
df = pd.read_csv('weather_data.csv')
```

```python
# Display the first few rows
print(df.head())

# Get information about the DataFrame
print(df.info())

# Get summary statistics
print(df.describe())
```

```python
import matplotlib.pyplot as plt

# To plot a histogram:
plt.hist(df['Humidity'], bins=20)
plt.xlabel('Humidity')
plt.ylabel('Frequency')
plt.title('Humidity Distribution')
plt.show()
```

```python
# This is a line plot

plt.plot(df['Wind'], df['Accumulation'])
plt.xlabel('Wind')
plt.ylabel('Accumulation')
plt.title('Wind Over Accumulation')
plt.show()
```

```python
#This is scatter plot

plt.scatter(df['Wind'], df['Humidity'])
plt.xlabel('Wind')
plt.ylabel('Humidity')
plt.title('Wind vs. Humidity')
plt.show()
```
